```python
# convert the data from dataframe to dictionary as mongo db is key value databases
data_dict = data.to_dict(orient="records")
```

```python
DB_name = "MLOps-day10" # user defined name
Collection_name = "MLOps-day10-cl" # user defined name
connection_stg = "replace with connection string"
```

```python
from pymongo import MongoClient

# Replace with your actual connection string
client = MongoClient(connection_stg)
db = client[DB_name]
collection = db[Collection_name]
# Project --> database --> collection , is the hierarchy in mongo db
```
